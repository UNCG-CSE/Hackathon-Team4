```python
import pandas as pd
import glob
```

```python
import warnings
warnings.filterwarnings('ignore')
import os
```

```python
path = r'C:\Users\shash\Downloads\Green_Hackathon-Fall2020-main\Green_Hackathon-Fall2020-main\data\Analysis' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for csv in all_files:
    cols = ['Actual','Predicted','obs_ci_lower','obs_ci_upper','Datetime']
    df = pd.read_csv(csv, index_col=None ,encoding = "ISO-8859-1",engine = 'python', sep=',', header=None,names=cols,error_bad_lines=False)
    df['Building_Name'] = os.path.basename(csv)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)



#for csv in all_files:
    #frame = pd.read_csv(csv)
    #frame['filename'] = os.path.basename(csv)
    #li.append(frame)
#frame
```
